```python
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import pandas as pd 
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
data_train_path='food_images/train'
data_test_path='food_images/test'
data_validation_path='food_images/validation'
```

```python
img_width=180
img_height=180
```

```python
image_path = '7.jpeg'
```

```python
image = load_img(image_path, target_size=(img_height, img_width))
image_array = img_to_array(image)
```

```python
data_train=tf.keras.utils.image_dataset_from_directory(
    data_train_path,
    shuffle=True,
    image_size=(img_width,img_height),
    batch_size=19,
    validation_split=False)
```

```python
data_cat=data_train.class_names
```

```python
data_cat
```

```python
data_val=tf.keras.utils.image_dataset_from_directory(
    data_validation_path,
    shuffle=True,
    image_size=(img_width,img_height),
    batch_size=19,
    validation_split=False)
```

```python
data_test=tf.keras.utils.image_dataset_from_directory(
    data_test_path,
    shuffle=True,
    image_size=(img_width,img_height),
    batch_size=19,
    validation_split=False)
```

```python
model = Sequential([
    layers.Rescaling(1./255),
    layers.Conv2D(16 , 3, padding='same' , activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32,3, padding='same', activation='relu'),
    layers.MaxPool2D(),
    layers.Conv2D(64, 3, padding= 'same', activation='relu'),
    layers.MaxPool2D(),
    layers.Flatten(),
    layers.Dense(128),
    layers.Dense(len(data_cat))
])
```

```python
model.compile(optimizer='adam', 
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
training_epochs= 200
history = model.fit(data_train, validation_data=data_val, epochs=training_epochs)
```

```python
image= '7.jpeg'
image= tf.keras.utils.load_img(image, target_size=(img_height, img_width))
img_arr= tf.keras.utils.array_to_img(image)
img_bat=tf.expand_dims(img_arr,0)
```

```python
predict=model.predict(img_bat)
```

```python
score=tf.nn.softmax(predict)
```

```python
print('food in image is {} with accuracy of {:0.2f}'.format(data_cat[np.argmax(score)],np.max(score)*100))
```
